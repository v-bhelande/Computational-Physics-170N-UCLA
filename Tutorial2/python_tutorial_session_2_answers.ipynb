```python
# Program 5
def fib_list(n: int):
    if n <= 0:
        return "n must be a positive integer! Exiting function..."
    if int(n) != n:
        return "n must be an integer! Exiting function..."
    if n == 1:
        return 1
    
    sequence = []
    num1 = 0
    num2 = 1
    sequence.append(num2)

    counter = 1
    while counter < n:
        sum = num1 + num2
        sequence.append(sum)
        num1 = num2
        num2 = sum
        counter += 1
    return sequence
 
# Test Cases
print(fib_list(0))
print(fib_list(-1.))
print(fib_list(2.5))
print(fib_list(1))
print(fib_list(2))
print(fib_list(5))
```

```python
# Program 6
def fib_n(n: int):
    # Get list of F_n
    F_n = fib_list(n)
    if type(F_n) is str:
        return "Invalid input!"
    if F_n == 1:
        return (F_n*F_n) + 1 - (1/F_n)
    
    # Transform elements
    for i in range(len(F_n)):
        F_n[i] = (F_n[i]*F_n[i]) + 1 - (1/F_n[i])
    return F_n

# Test Cases
print(fib_n(0))
print(fib_n(-1.))
print(fib_n(2.5))
print(fib_n(1))
print(fib_n(2))
print(fib_n(5))
```
