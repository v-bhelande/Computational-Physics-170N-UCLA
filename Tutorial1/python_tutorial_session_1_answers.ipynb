### Answers -- writing functions

Create your new cell(s) below this one.

```python
# Program 1
def my_fifth_power(num):
    return num*num*num*num*num

# Test Cases
print(my_fifth_power(0))
print(my_fifth_power(-1))
print(my_fifth_power(2))
```

```python
# Program 2
def my_power(num, pow):
    if pow < 0:
        return "Power must be a non-negative number! Exiting function..."
    if int(pow) != pow:
        return "Power must be an integer! Exiting function..."
    result = 1
    for i in range(pow):
        result *= num
    return result

# Test Cases
print(my_power(100, 0))
print(my_power(10, -1))
print(my_power(1, 2.5))
print(my_power(-1, 9))
print(my_power(2, 5))
```

```python
# Program 3
def my_absolute(num: float):
    if num < 0:
        return -1. * num
    else:
        return num
    
# Test Cases
print(my_absolute(-9))
print(my_absolute(-1.))
print(my_absolute(-0.5))
print(my_absolute(0.002))
print(my_absolute(0))
```

```python
# Program 4
def my_max(num1, num2):
    if num1 >= num2:
        return num1
    else:
        return num2

# Test Cases
print(my_max(2, 2))
print(my_max(2.00000001, 2.))
print(my_max(-2.7, -3.1))
print(my_max(1., 0.))
```

```python
# Program 5
def my_max_three(num1, num2, num3):
    if num1 >= num2 and num1 >= num3:
        return num1
    elif num2 >= num1 and num2 >= num3:
        return num2
    else:
        return num3
    
# Test Cases
print(my_max_three(0., -0., 0.))
print(my_max_three(1.00001, 1.0001, 1.001))
print(my_max_three(-1.00001, -1.0001, -1.001))
print(my_max_three(-1., -2., -3.))
print(my_max_three(2., 3., 5.))
```

```python
# Program 6
def my_integer_sum(n: int):
    if int(n) != n:
        return "Number must be an integer! Exiting function..."
    if n <= 0:
        return "Integer must be positive! Exiting function..."
    sum = 0
    for i in range (1, n+1):
        sum += i
    return sum

# Test Cases
print(my_integer_sum(2.4))
print(my_integer_sum(-1))
print(my_integer_sum(0))
print(my_integer_sum(5))
print(my_integer_sum(11))
```

```python
# Program 7
def my_integer_sum_power(n:int, p=1):
    if int(n) != n:
        return "Number must be an integer! Exiting function..."
    if n <= 0:
        return "Integer must be positive! Exiting function..."
    if int(p) != p:
        return "Power must be an integer! Exiting function..."
    if p <= 0:
        return "Power must be positive! Exiting function..."
    cum_sum = 0
    for i in range(1, n+1):
        cum_sum += my_power(i, p)
    return cum_sum

# Test Cases
print(my_integer_sum_power(-1))
print(my_integer_sum_power(1.1))
print(my_integer_sum_power(1, 0))
print(my_integer_sum_power(1, 1.1))
print(my_integer_sum_power(2))
print(my_integer_sum_power(3, 2))
print(my_integer_sum_power(5, 4))
```
