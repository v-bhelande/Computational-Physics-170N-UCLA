### Problem 1.4 - The Logistic Map

```python
import numpy as np
import matplotlib.pyplot as plt
from python_problem_set import logistic

n_list = np.arange(0, 101)

x0 = 0.5
r_list = [0.5, 1.5, 2.5, 3.3, 3.5]
x_r1 = [logistic(i, r_list[0], x0) for i in range(101)]
x_r2 = [logistic(i, r_list[1], x0) for i in range(101)]
x_r3 = [logistic(i, r_list[2], x0) for i in range(101)]
x_r4 = [logistic(i, r_list[3], x0) for i in range(101)]
x_r5 = [logistic(i, r_list[4], x0) for i in range(101)]

# 1. r = 0.5
plt.plot(n_list, x_r1)
plt.xlabel("n")
plt.ylabel("$x_{n+1}$")
plt.title("Logistic Plot for r = 0.5")
plt.show()

# 2. r = 1.5
plt.plot(n_list, x_r2)
plt.xlabel("n")
plt.ylabel("$x_{n+1}$")
plt.title("Logistic Plot for r = 1.5")
plt.show()

# 3. r = 2.5
plt.plot(n_list, x_r3)
plt.xlabel("n")
plt.ylabel("$x_{n+1}$")
plt.title("Logistic Plot for r = 2.5")
plt.show()

# 4. r = 3.3
plt.plot(n_list, x_r4)
plt.xlabel("n")
plt.ylabel("$x_{n+1}$")
plt.title("Logistic Plot for r = 3.3")
plt.show()

# 5. r = 3.5
plt.plot(n_list, x_r5)
plt.xlabel("n")
plt.ylabel("$x_{n+1}$")
plt.title("Logistic Plot for r = 3.5")
plt.show()

# Plot all of them together
plt.plot(n_list, x_r1, label="r = 0.5")
plt.plot(n_list, x_r2, label="r = 1.5")
plt.plot(n_list, x_r3, label="r = 2.5")
plt.plot(n_list, x_r4, label="r = 3.3")
plt.plot(n_list, x_r5, label="r = 3.5")
plt.xlabel("n")
plt.ylabel("$x_{n+1}$")
plt.title("Logistic Plot as a Function of r")
plt.legend(loc="upper right")
plt.show()
```

### Problem 2.4 - A Physical Application: Projectile Range Maximization

#### a.

r(v, $\phi$, $\theta$) = $\frac{vsin(2(\phi+\theta))}{gcos\phi}$

####  b.

$\theta_{max}$ = $\frac{\pi}{4}$ - $\phi$

#### c.

Going to assume v = 1 and $\theta$ = 0 degrees for simplicity.

**Approach:** Plug value of $\phi$ into equation for range from part 1. Compute derivative wrt $\theta$ by hand and define function in code. Since we expect a proportionate relationship between v and r, we will find the maximum range for a given $\theta$ instead and use the function from Problem 2.3 to find the root. 

```python
from python_problem_set import phi_0
from python_problem_set import phi_pi_8
from python_problem_set import phi_pi_4
from python_problem_set import phi_3pi_8

r1 = bisection_root(phi_0, 0, np.pi/2, 1e-12)
print("Maximum range at phi = 0 occurs at theta =", r1*(180/np.pi), "degrees")
# This agrees with with the numerical result of 45.0 degrees!

r2 = bisection_root(phi_pi_8, 0, np.pi/2, 1e-9)
print("Maximum range at phi = pi/8 occurs at theta =", r2*(180/np.pi), "degrees")
# This agrees with with the numerical result of 22.5 degrees!

r3 = bisection_root(phi_pi_4, 0, np.pi/2, 1e-9)
print("Maximum range at phi = pi/4 occurs at theta =", r3*(180/np.pi), "degrees")
# This agrees with with the numerical result of 0.0 degrees!

r4 = bisection_root(phi_3pi_8, 0, np.pi/2, 1e-9)
print("Maximum range at phi = 3*pi/8 occurs at theta =", r4*(180/np.pi), "degrees")
# This agrees with with the numerical result of 67.5 degrees!
```
